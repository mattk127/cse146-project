```python
test_df = iowa_df[list1]


race={'Black - Non-Hispanic':1,
      'White - Non-Hispanic':2,
      'American Indian or Alaska Native - Non-Hispanic':3,
      'White - Hispanic':4,
      'American Indian or Alaska Native - Hispanic':5,
      'Asian or Pacific Islander - Non-Hispanic':6,
      'White -':7,
      'Black - Hispanic':8,
      'Black -':9,
      'Asian or Pacific Islander - Hispanic':10,
      'N/A -':0,
      np.nan:0
     }
age={'Under 25':1,
     '25-34':2, 
     '35-44':3, 
     '45-54':4,  
     '55 and Older':5, 
     np.nan:0
    }
sex={'Male':1, 
     'Female':2,
     np.nan:0
    }
offenseC={'A Felony':1,
         'B Felony':2,
         'C Felony':3,
         'D Felony':4,
         'Other Felony':5,
         'Felony - Enhanced':6,
         'Other Felony (Old Code)':7,
         'Felony - Mandatory Minimum':8,
         'Felony - Enhancement to Original Penalty':9,
         'Simple Misdemeanor':10,
         'Aggravated Misdemeanor':11,
         'Serious Misdemeanor':12,
         'Sexual Predator Community Supervision':13,
         'Other Misdemeanor':14,
         'Special Sentence 2005':15
        }
offenceT={'Violent':1, 
          'Property':2, 
          'Public Order':3,
          'Drug':4, 
          'Other':0}
recidivism={'Yes':1,
            'No':0}
test_df.columns=['Race','Age','Sex','Classification','Type','Recidivism']
test_df.Race = [race[item] for item in test_df.Race] 
test_df.Age = [age[item] for item in test_df.Age]
test_df.Sex = [sex[item] for item in test_df.Sex] 
test_df.Classification = [offenseC[item] for item in test_df.Classification] 
test_df.Type = [offenceT[item] for item in test_df.Type] 
test_df.Recidivism = [recidivism[item] for item in test_df.Recidivism] 
#alist
# test_df
list_u25 = []
list_25_34 = []
list_35_44 = []
list_45_54 = []
list_55up = []
for row in test_df['Age']:
    if row == 1:
        list_u25.append(1)
        list_25_34.append(0)
        list_35_44.append(0)
        list_45_54.append(0)
        list_55up.append(0)
    elif row == 2:
        list_u25.append(0)
        list_25_34.append(1)
        list_35_44.append(0)
        list_45_54.append(0)
        list_55up.append(0)
    elif row == 3:
        list_u25.append(0)
        list_25_34.append(0)
        list_35_44.append(1)
        list_45_54.append(0)
        list_55up.append(0)
    elif row == 4:
        list_u25.append(0)
        list_25_34.append(0)
        list_35_44.append(0)
        list_45_54.append(1)
        list_55up.append(0)  
    elif row == 5:
        list_u25.append(0)
        list_25_34.append(0)
        list_35_44.append(0)
        list_45_54.append(0)
        list_55up.append(1)  
    else:
        list_u25.append(0)
        list_25_34.append(0)
        list_35_44.append(0)
        list_45_54.append(0)
        list_55up.append(0)
        

blist = []
wlist = []
alist = []
hlist = [] 
olist = []
for row in test_df['Race']:
    if row == 1 or row == 8 or row == 9:
        blist.append(1)
        wlist.append(0)
        alist.append(0)
        hlist.append(0)
        olist.append(0)
    elif row == 2 or row == 7:
        blist.append(0)
        wlist.append(1)
        alist.append(0)
        hlist.append(0)
        olist.append(0)
    elif row == 4 or row == 5:
        blist.append(0)
        wlist.append(0)
        alist.append(0)
        hlist.append(1)
        olist.append(0)
    elif row == 6 or row == 10:
        blist.append(0)
        wlist.append(0)
        alist.append(1)
        hlist.append(0)
        olist.append(0)
    elif row == 3 or row == 0:
        blist.append(0)
        wlist.append(0)
        alist.append(0)
        hlist.append(0)
        olist.append(1)

        
male = []
female = []
for row in test_df['Sex']:
    if row == 1:
        male.append(1)
        female.append(0)
    elif row == 2:
        male.append(0)
        female.append(1)
    else:
        male.append(0)
        female.append(0)
        

violent_l = []
prop_l = []
drug_l = []
pub_ord = []
other_l = []
for row in test_df['Type']:
    if row == 0:
        violent_l.append(0)
        prop_l.append(0)
        drug_l.append(0)
        pub_ord.append(0)
        other_l.append(1)
    elif row == 1:
        violent_l.append(1)
        prop_l.append(0)
        drug_l.append(0)
        pub_ord.append(0)
        other_l.append(0)
    elif row == 2:
        violent_l.append(0)
        prop_l.append(1)
        drug_l.append(0)
        pub_ord.append(0)
        other_l.append(0)
    elif row == 3:
        violent_l.append(0)
        prop_l.append(0)
        drug_l.append(0)
        pub_ord.append(1)
        other_l.append(0)
    elif row == 4:
        violent_l.append(0)
        prop_l.append(0)
        drug_l.append(1)
        pub_ord.append(0)
        other_l.append(0) 
        
        
        
        
top_fel = []
bot_fel = []
mis = []
other = []
for row in test_df['Classification']:
    if row == 1 or row == 2 or row == 3:
        top_fel.append(1)
        bot_fel.append(0)
        mis.append(0)
        other.append(0)
    elif row == 4 or row == 5 or row == 6 or row == 7 or row == 8 or row == 9:
        top_fel.append(0)
        bot_fel.append(1)
        mis.append(0)
        other.append(0)
    elif row == 10 or row == 11 or row == 12 or row == 13 or row == 14:
        top_fel.append(0)
        bot_fel.append(0)
        mis.append(1)
        other.append(0)
    elif row == 15:
        top_fel.append(0)
        bot_fel.append(0)
        mis.append(0)
        other.append(1)
        
test_df.insert(1,"White",wlist,True)
test_df.insert(2,"Black",blist,True)
test_df.insert(3,"Hispanic",hlist,True)
test_df.insert(4,"Asian",alist,True)
test_df.insert(5,"Other",olist,True)
test_df.insert(6,"Under 25",list_u25,True)
test_df.insert(7,"25-34",list_25_34,True)
test_df.insert(8,"35-44",list_35_44,True)
test_df.insert(9,"45-54",list_45_54,True)
test_df.insert(10,"55+",list_55up,True)
test_df.insert(11,"Male",male,True)
test_df.insert(12,"Female",female,True)
test_df.insert(13,"Serious Felony",top_fel,True)
test_df.insert(14,"Felony",bot_fel,True)
test_df.insert(15,"Misdemeanor",mis,True)
test_df.insert(16,"Other Offense",other,True)
test_df.insert(17,"Violent",violent_l,True)
test_df.insert(18,"Property",prop_l,True)
test_df.insert(19,"Public Order",pub_ord,True)
test_df.insert(20,"Drug",drug_l,True)
test_df.insert(21,"Other",other_l,True)
del test_df['Race']
del test_df['Age']
del test_df['Sex']
del test_df['Classification']
del test_df['Type']
test_df
```
