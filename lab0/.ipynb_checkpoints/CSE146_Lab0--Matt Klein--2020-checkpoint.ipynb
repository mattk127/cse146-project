### Problem 7 (15 points)

Plot the _average life expectancy_ for _all_ countries as a function of year.  

```python
import math
lifeexpectancy.loc[:,'life expectancy at birth (both sexes)'] = lifeexpectancy.loc[:,'life expectancy at birth (both sexes)'].astype(float)
lifeexpectancy.loc[:,'year'] = lifeexpectancy.loc[:,'year'].astype(float)

a = lifeexpectancy.loc[(lifeexpectancy['year']== 2000.0)]
b = lifeexpectancy.loc[(lifeexpectancy['year']== 2001.0)]
c = lifeexpectancy.loc[(lifeexpectancy['year']== 2002.0)]
d = lifeexpectancy.loc[(lifeexpectancy['year']== 2003.0)]
e = lifeexpectancy.loc[(lifeexpectancy['year']== 2004.0)]
f = lifeexpectancy.loc[(lifeexpectancy['year']== 2005.0)]
g = lifeexpectancy.loc[(lifeexpectancy['year']== 2006.0)]
h = lifeexpectancy.loc[(lifeexpectancy['year']== 2007.0)]
i = lifeexpectancy.loc[(lifeexpectancy['year']== 2008.0)]
j = lifeexpectancy.loc[(lifeexpectancy['year']== 2009.0)]
k = lifeexpectancy.loc[(lifeexpectancy['year']== 2010.0)]
l = lifeexpectancy.loc[(lifeexpectancy['year']== 2011.0)]
m = lifeexpectancy.loc[(lifeexpectancy['year']== 2012.0)]
n = lifeexpectancy.loc[(lifeexpectancy['year']== 2013.0)]
o = lifeexpectancy.loc[(lifeexpectancy['year']== 2014.0)]
p = lifeexpectancy.loc[(lifeexpectancy['year']== 2015.0)]


counta = 0
countb = 0
countc = 0
countd = 0
counte = 0
countf = 0
countg = 0
counth = 0
counti = 0
countj = 0
countk = 0
countl = 0
countm = 0
countn = 0
counto = 0
countp = 0

suma = 0.0
sumb = 0.0
sumc = 0.0
sumd = 0.0
sume = 0.0
sumf = 0.0
sumg = 0.0
sumh = 0.0
sumi = 0.0
sumj = 0.0
sumk = 0.0
suml = 0.0
summ = 0.0
sumn = 0.0
sumo = 0.0
sump = 0.0

years = [2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015]
avg = []
for row in a.itertuples():
    suma = suma + row._3
    counta = counta + 1
avg.append(round(suma/counta,2))

for row in b.itertuples():
    sumb = sumb + row._3
    countb = countb + 1
avg.append(round(sumb/countb,2))

for row in c.itertuples():
    sumc = sumc + row._3
    countc = countc + 1
avg.append(round(sumc/countc,2))

for row in d.itertuples():
    sumd = sumd + row._3
    countd = countd + 1
avg.append(round(sumd/countd,2))

for row in e.itertuples():
    sume = sume + row._3
    counte = counte + 1
avg.append(round(sume/counte,2))

for row in f.itertuples():
    sumf = sumf + row._3
    countf = countf + 1
avg.append(round(sumf/countf,2))

for row in g.itertuples():
    sumg = sumg + row._3
    countg = countg + 1
avg.append(round(sumg/countg,2))

for row in h.itertuples():
    sumh = sumh + row._3
    counth = counth + 1
avg.append(round(sumh/counth,2))

for row in i.itertuples():
    sumi = sumi + row._3
    counti = counti + 1
avg.append(round(sumi/counti,2))

for row in j.itertuples():
    sumj = sumj + row._3
    countj = countj + 1
avg.append(round(sumj/countj,2))

for row in k.itertuples():
    sumk = sumk + row._3
    countk = countk + 1
avg.append(round(sumk/countk,2))

for row in l.itertuples():
    suml = suml + row._3
    countl = countl + 1
avg.append(round(suml/countl,2))

for row in m.itertuples():
    summ = summ + row._3
    countm = countm + 1
avg.append(round(summ/countm,2))

for row in n.itertuples():
    if math.isnan(row._3):
        continue
    else:
        sumn = sumn + row._3
        countn = countn + 1
avg.append(round(sumn/countn,2))

for row in o.itertuples():
    sumo = sumo + row._3
    counto = counto + 1
avg.append(round(sumo/counto,2))

for row in p.itertuples():
    sump = sump + row._3
    countp = countp + 1
avg.append(round(sump/countp,2))

#print out data as scatter plot
plt.plot(years, avg,'o')
plt.title('Average Life Expecatancy for All Countries')
```
